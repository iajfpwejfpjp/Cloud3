```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Books = pd.read_csv("D:/Cloud/archive/books.csv")
Books = Books.dropna()
Books.drop(['work_id', 'isbn', 'isbn13', 'image_url', 'small_image_url'], axis=1, inplace=True)
harry_potter_df = Books[Books['title'].str.contains("Harry Potter")]
harry_potter_df
```

```python
most_selling_books = harry_potter_df.sort_values(by='books_count', ascending=False).head(5)
print("Most Selling Harry Potter Books:")
print(most_selling_books[['title', 'books_count']])
```

```python
average_rating = harry_potter_df['average_rating'].mean()
print("Average Rating of Harry Potter Books:", average_rating)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(harry_potter_df['average_rating'], bins=20, kde=True, color='skyblue')
plt.title('Average Rating Distribution of Harry Potter Books')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```
